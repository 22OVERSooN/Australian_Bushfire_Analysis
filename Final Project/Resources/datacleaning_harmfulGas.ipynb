```python
df_so2 = pd.read_excel("so2_pm2.5.xls",sheet_name='so2')
```

```python
df_so2
```

```python
station =[]
for col in df_so2.columns:
    if col != 'Date':
        station.append(col)
    else: 
        pass
len(station)    
```

```python
df_so2['RANDWICK SO2 monthly average [pphm]'].append(df_so2['ROZELLE SO2 monthly average [pphm]'],ignore_index=True).append(df_so2['LINDFIELD SO2 monthly average [pphm]'],ignore_index=True)
```

```python
df_so2['ROZELLE SO2 monthly average [pphm]'].values.tolist()
```

```python

df1 = df_so2['RANDWICK SO2 monthly average [pphm]']
df2 = df_so2['Date']
for i in range(1,39):
    df1=df1.append(df_so2[station[i]],ignore_index=True)
    df2=df2.append(df_so2['Date'],ignore_index=True)
        
```

```python
df3 = pd.DataFrame({'period':df2,'so2_avg':df1})

df3
```

```python
df4 = pd.DataFrame({'p':df_so2['Date']})
```

```python
stations =[]

for i in range(0,len(station)):
    stations.append(station[i][:-27])

stations
```

```python


df5 = pd.concat([df4.assign(station=s) for s in stations],ignore_index=True)

df3['station']=df5['station'].copy()

df3
```

```python
df_reg = pd.read_excel(r"regions.xlsx",engine='openpyxl')

df_reg
```

```python
df_reg['station']=df_reg['station'].str.upper()

df_reg
```

```python
df_com = pd.merge(df3,df_reg,on='station',how='inner')

df_com
```

```python
df_pm = pd.read_excel("so2_pm2.5.xls",sheet_name='pm2.5')


df_pm
```

```python
station2 =[]
for col in df_pm.columns:
    if col != 'Date':
        station2.append(col)
    else: 
        pass

df6 = df_pm['RANDWICK PM2.5 monthly average']
df7 = df_pm['Date']
for i in range(1,51):
    df6=df6.append(df_pm[station2[i]],ignore_index=True)
    df7=df7.append(df_pm['Date'],ignore_index=True)

df8 = pd.DataFrame({'period':df7,'pm2.5_avg':df6})

df9 = pd.DataFrame({'p':df_pm['Date']})

stations2 =[]

for i in range(0,len(station2)):
    stations2.append(station2[i][:-22])

df10 = pd.concat([df9.assign(station=s) for s in stations2],ignore_index=True)

df8['station']=df10['station'].copy()

df8
```

```python
df0=df_com.join(df8['pm2.5_avg'])

df0
```

```python
dfc = df0.loc[(df0['station'] != 'LINDFIELD') & 
              (df0['station'] != 'KEMBLA GRANGE') &
             (df0['station'] != 'ST MARYS') &
             (df0['station'] != 'VINEYARD') &
             (df0['station'] != 'MACARTHUR') &
             (df0['station'] != 'OAKDALE') &
             (df0['station'] != 'MERRIWA') &
             (df0['station'] != 'GUNNEDAH') &
             (df0['station'] != 'PORT MACQUARIE') &
             (df0['station'] != 'ROUSE HILL') &
             (df0['station'] != 'LIDCOMBE') &
             (df0['station'] != 'PENRITH') &
             (df0['station'] != 'MORISSET')].reset_index(drop=True)
dfc
```

```python
dfd=dfc.fillna(method='ffill')
```

```python
dfd= dfd.rename(columns={'pm2.5_avg':'pm25_avg'})

dfd
```

```python
dfd.to_csv('test1.csv',index=False)
```
