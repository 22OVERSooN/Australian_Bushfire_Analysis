```python
import csv
import pandas as pd
import numpy as np
import datetime as dt
```

## Read the raw data from csv

```python
#bushfire data 2019 from Nasa
bushfire_2019_raw = pd.read_csv("Resources/Raw/modis_2019_Australia.csv")
bushfire_2019_raw.head()
```

```python
#bushfire data 2020 from Nasa
bushfire_2020_raw = pd.read_csv("Resources/Raw/modis_2020_Australia.csv")
bushfire_2020_raw.head()
```

```python
#Weather data May/2019-June/2020 from BOM
aus_weather_raw = pd.read_csv("Resources/Raw/Australia_weather.csv")
aus_weather_raw
```

```python
#Weather station data from BOM
weather_station_raw = pd.read_csv("Resources/Raw/weather_station.csv")
weather_station_raw
```

## Clean the Bushfire Dataset

```python
#change column name 
bushfire_2019_new = bushfire_2019_raw.rename(columns = {'acq_date':'date'})
bushfire_2020_new = bushfire_2020_raw.rename(columns = {'acq_date':'date'})
```

```python
#choose useable column
bushfire_2019 = bushfire_2019_new[['latitude','longitude','brightness','date']]
bushfire_2020 = bushfire_2020_new[['latitude','longitude','brightness','date']]
bushfire_2020.head()
```

```python
#combine two dataset into one
bushfire_19_20 = pd.concat([bushfire_2019,bushfire_2020])
bushfire_19_20 = bushfire_19_20.dropna()
bushfire_19_20.reset_index(drop = True)
bushfire_19_20
```

```python
bushfire_19_20['date'] = pd.to_datetime(bushfire_19_20['date'])
bushfire_19_20.to_csv("Resources/Bushfire_2019&2020.csv")
```

```python
#fit the data of the weather dataset
bushfire_match = bushfire_19_20[(bushfire_19_20['date']>'2019-05-01')&(bushfire_19_20['date']<'2020-06-30')]
bushfire_match.to_csv("Resources/Bushfire_05-2019_06-2020.csv")
bushfire_match
```

## Clean the weather dataset

```python
aus_weather_raw.info()
```

```python
#change column name
aus_weather_new = aus_weather_raw.rename(columns = {'Min °C':'Min Temp(°C)','Max °C':'Max Temp(°C)','Max Wind gust Dir':"Max Wind Dir",'Max wind gust Spd - km/h':"Max Wind Spd(km/h)","Day":"Day1",'Date':'Day'})

aus_weather_new.loc[aus_weather_new['Month']=='May','Month']=5
aus_weather_new.loc[aus_weather_new['Month']=='June','Month']=6
aus_weather_new.loc[aus_weather_new['Month']=='July','Month']=7
aus_weather_new.loc[aus_weather_new['Month']=='August','Month']=8
aus_weather_new.loc[aus_weather_new['Month']=='September','Month']=9
aus_weather_new.loc[aus_weather_new['Month']=='October','Month']=10
aus_weather_new.loc[aus_weather_new['Month']=='November','Month']=11
aus_weather_new.loc[aus_weather_new['Month']=='December','Month']=12
aus_weather_new.loc[aus_weather_new['Month']=='January','Month']=1
aus_weather_new.loc[aus_weather_new['Month']=='February','Month']=2
aus_weather_new.loc[aus_weather_new['Month']=='March','Month']=3
aus_weather_new.loc[aus_weather_new['Month']=='April','Month']=4
aus_weather_new['Month'] = pd.to_numeric(aus_weather_new['Month'])

aus_weather_new.info()
#'May', 'June', 'July', 'August', 'September', 'October','November', 'December', 'January', 'February', 'March', 'April'
```

```python
#filter the column 
aus_weather = aus_weather_new[['State','Station','Min Temp(°C)','Max Temp(°C)','Rain(mm)','Max Wind Dir','Max Wind Spd(km/h)']]
aus_weather
```

```python
#split the station name
aus_weather_final = aus_weather.copy()
aus_weather_final['Station'] = aus_weather['Station'].str.rsplit("{", expand = True, n = 1)[0]
aus_weather_final
```

```python
#split the station no.
aus_weather_final['Station No.'] = aus_weather['Station'].str.rsplit(" 0", expand = True, n = 1)[1]
aus_weather_final['Station No.'] = aus_weather_final['Station No.'].str.rsplit("}", expand = True, n = 1)[0]
aus_weather_final
```

```python
#datetime the three columns of year,month,day
aus_weather_final['Date'] = pd.to_datetime(aus_weather_new[["Year", "Month", "Day"]])
aus_weather_final
```

```python
#export the final csv file
aus_weather_final = aus_weather_final[['Date','State','Station','Station No.','Min Temp(°C)','Max Temp(°C)','Rain(mm)','Max Wind Dir','Max Wind Spd(km/h)']]
aus_weather_final.to_csv("Resources/weather_Australia.csv")
```

## Merge the station data with weather data 

```python
#merge the two dataset table on the station number
aus_weather_final["Station No."] = pd.to_numeric(aus_weather_final["Station No."])
aus_weather_with_station = aus_weather_final.merge(weather_station_raw,left_on="Station No.",right_on='STN_NUM')

aus_weather_with_station
```

```python
#clean the whole dataset
aus_weather_with_station = aus_weather_with_station.rename(columns = {'LATITUDE':'latitude','LONGITUDE':'longitude'})
aus_weather_with_station_final = aus_weather_with_station[['Date','State','Station','Station No.','latitude','longitude','Min Temp(°C)','Max Temp(°C)','Rain(mm)','Max Wind Dir','Max Wind Spd(km/h)']]
aus_weather_with_station_final = aus_weather_with_station_final.dropna()
aus_weather_with_station_final
```

```python
#to csv file
aus_weather_with_station_final.to_csv("Resources/weather_Australia_with_station.csv")
```

```python
# Generate a summary statistics table for each state over a year

grouped_weatcher=aus_weather_with_station_final.groupby(['Date','State'])
grouped_weatcher_temp=grouped_weatcher['Max Temp(°C)'].mean()
grouped_weatcher_rain=grouped_weatcher['Rain(mm)'].mean()
```

```python
summary_stats= pd.DataFrame({"Avg Max Temp": grouped_weatcher_temp,"Avg Rain mm":grouped_weatcher_rain})
summary_stats.reset_index()
```

```python
summary_stats.to_csv("Resources/weather_summary_stats.csv")
```
